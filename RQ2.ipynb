```python
import pandas as pd
```

```python
# Read the RQ2 database, that we did specially for these exrecise
RQ2 = pd.read_csv('RQ2.csv')
cols_of_interest = ['Hour', 'passenger_count', 'Borough']
RQ2 = RQ2[cols_of_interest]
```

```python
# Total number of passengers per hour overall New York
passengers_per_hour = RQ2.groupby(["Hour"])['passenger_count'].sum()
```

```python
print(passengers_per_hour)
```

```python
passengers_per_hour.plot(title="Passengers per hour overall New York")
```

```python
# Average number of passengers, per hour per taxi overall New York
passengers_per_hour_per_taxi = RQ2.groupby(["Hour"])['passenger_count'].mean()
```

```python
passengers_per_hour_per_taxi.plot(title="Average number of passengers per hour per taxi overall New York")
```
