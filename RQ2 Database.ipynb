```python
import pandas as pd
```

```python
# Import all the data bases and select only the columns that I need for the exercise
```

```python
cols_of_interest = ['tpep_pickup_datetime', 'passenger_count', 'PULocationID']
```

```python
jan = pd.read_csv('yellow_tripdata_2018-01.csv')
jan = jan[cols_of_interest]
```

```python
feb = pd.read_csv('yellow_tripdata_2018-02.csv')
feb = feb[cols_of_interest]
```

```python
mar = pd.read_csv('yellow_tripdata_2018-03.csv')
mar = mar[cols_of_interest]
```

```python
apr = pd.read_csv('yellow_tripdata_2018-04.csv')
apr = apr[cols_of_interest]
```

```python
may = pd.read_csv('yellow_tripdata_2018-05.csv')
may = may[cols_of_interest]
```

```python
jun = pd.read_csv('yellow_tripdata_2018-06.csv')
jun  = jun[cols_of_interest]
```

```python
# Concatenate all this data bases in one
frames = [jan, feb, mar, apr, may, jun]
jan_to_jun = pd.concat(frames)
```

```python
# Transform the variable tpep_pickup_datetime in datetime format
jan_to_jun['tpep_pickup_datetime'] = pd.to_datetime(jan_to_jun['tpep_pickup_datetime'])
```

```python
# And create a column with the hour 
jan_to_jun["Hour"] = jan_to_jun.tpep_pickup_datetime.dt.hour
```

```python
# Merge the jan_to_jun data base with taxi zone data
```

```python
# Change the name variable 'PULocationID' to 'LocationID' in the jan_to_jun data base
jan_to_jun = jan_to_jun.rename(columns={'PULocationID': 'LocationID'})
```

```python
# Import the taxi_zone_lookup data base
taxi_zone_data = pd.read_csv('taxi_zone_lookup.csv')
```

```python
# Merge them and select only the columns of interest
RQ2 = pd.merge(jan_to_jun, taxi_zone_data, on="LocationID")
cols_of_interest = ['Hour', 'passenger_count', 'Borough']
RQ2 = RQ2[cols_of_interest]
```

```python
# Download the database
RQ2.to_csv('RQ2.csv')
```
